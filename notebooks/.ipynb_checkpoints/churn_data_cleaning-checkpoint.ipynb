```python
import pandas as pd

df = pd.read_csv(r"C:\Data Analyst Projects\customer-churn\data\Telco-Customer-Churn.csv")
df.head()
```

```python
df.drop(['customerID'], axis=1, inplace=True, errors='ignore')
```

```python
if 'TotalCharges' in df.columns:
    df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
else:
    print("⚠️ TotalCharges column not found. Creating from MonthlyCharges * tenure...")
    df['TotalCharges'] = df['MonthlyCharges'] * df['tenure']
```

```python
df['Churn'] = df['Churn'].str.strip().map({'Yes': 1, 'No': 0})
```

```python
cat_cols = df.select_dtypes(include=['object']).columns
print(cat_cols)
```

```python
binary_cols = ['Partner','Dependents','PhoneService','PaperlessBilling']
for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

```python
df = pd.get_dummies(df, columns=['InternetService','PaymentMethod','Contract'], drop_first=True)
```

```python
df['MonthlyCostPerService'] = df['MonthlyCharges'] / (df['tenure']+1)

```

```python
df.to_csv("../data/churn_cleaned.csv", index=False)
```

```python
#Average Monthly Spend per Tenure
df['AvgMonthlySpend'] = df['TotalCharges'] / (df['tenure'] + 1)
```
